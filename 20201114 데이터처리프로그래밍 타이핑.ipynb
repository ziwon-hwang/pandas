#비율 기준 pd.qcut, 구간갯수는 1이 넘어갈수 있으나

```python
pd.qcut(df5.A,3)
```

```python
pd.qcut(df5.A,3).value_counts()
```

```python
pd.qcut(df5.A,[0,4,7,10]) #1넘어가면 오류남
```

```python
pd.qcut(df5.A,[0,0.4,0.7,1])
```

```python
pd.qcut(df5.A,[0,0.4,0.7,1],labels=['L','M','H'])
```
