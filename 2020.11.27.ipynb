pivot이 안되는 경우(피벗은 본데이터에서)

1.중복 값이 있는 경우
2.columns가 2개이상인 경우
3.index가 2개 이상인 경우(에를들어 multi-index)

```python
df2.pivot(index='origin',columns='model_year',values='cylinders')
```

```python
df2.pivot(index='origin',columns=['model_year','mpg'],values='cylinders')
```

```python
df2.pivot(index=['origin','mpg'],columns='model_year',values='cylinders')
```

df2.pivot_table( values=None,
    index=None,
    columns=None,
    aggfunc='mean')

```python
df2.pivot_table(index='origin',columns='model_year',values='cylinders') #그룹바이는 값별 총평균값이나 총합계등을 이용하여 피벗테이블값과 다름
```

```python
df2.pivot_table(index='origin',columns='model_year',values='cylinders',aggfunc='median')
```

```python
df2.pivot_table(index='origin',columns='model_year',values='cylinders',aggfunc='sum')
```

```python
df2.pivot_table(index='origin',columns=['model_year','mpg'],values='cylinders')
```

```python
df2.pivot_table(index=['origin','mpg'],columns='model_year',values='cylinders')
```
